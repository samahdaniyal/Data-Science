```python
df = pd.read_csv('formatted_580SurveyCleanup2.csv')
df.head()
```

```python
# Checking for missing values in each column
missing_values = df.isnull().sum()

# Listing columns with missing values
missing_columns = missing_values[missing_values > 0]
print("Columns with missing values and number of values missing:\n", missing_columns)
```

```python
#dropping access code and email address columns because they have no values hence these columns do not affect the data in any manner.
df.drop(columns=['Access Code', 'Email Address'], inplace=True) 
```

```python
#using reverse geocoding to fill in the missing values of region and city/town/district, reference: https://geopy.readthedocs.io/en/latest/
from geopy.geocoders import Nominatim

# Function to reverse geocode
def reverse_geocode(lat, lon):
    geolocator = Nominatim(user_agent="DS_HW")
    try:
        # Skip if latitude or longitude is NaN
        if pd.isna(lat) or pd.isna(lon):
            return '', ''
        
        # Check if latitude and longitude seem swapped, Check if latitude and longitude values are out of range Latitude should be between -90 and 90, and longitude between -180 and 180
        if abs(lat) > 90 or abs(lon) > 180:
            lat, lon = lon, lat  # Swap them
        # Use the geolocator to reverse geocode the coordinates
        # 'language' specifies the language for the returned address, 'addressdetails=True' requests detailed address information
        location = geolocator.reverse((lat, lon), language='en', addressdetails=True)
        if location:
            address = location.raw.get('address', {})
            city = address.get('city', '') or address.get('town', '') or address.get('village', '') # Get the city, town, or village, preferring 'city' over 'town' or 'village'
            region = address.get('state', '') or address.get('region', '') # Get the state or region, preferring 'state' over 'region'
            return city, region
        return '', ''
    except Exception as e:
        print(f"Error: {e} for coordinates lat={lat}, lon={lon}")
        return '', ''  # Always return two empty strings in case of error

# Update only NaN values using reverse geocoding
def update_missing_info(row):
    if pd.isna(row['Region']) or pd.isna(row['City / Town / District']):
        city, region = reverse_geocode(row['Latitude'], row['Longitude'])
        return pd.Series([city, region], index=['City / Town / District', 'Region'])
    else:
        return pd.Series([row['City / Town / District'], row['Region']], index=['City / Town / District', 'Region'])

df[['City / Town / District', 'Region']] = df.apply(update_missing_info, axis=1)
```

```python
#if even after geocoding Region or city is Null (geocoding code above would make it '' instead of Nan) filling it with "Not Available" as that data hasn't been found or given
def fill_missing(value):
    return 'Not Available' if pd.isna(value) or value == '' else value

df['Region'] = df['Region'].apply(fill_missing)
df['City / Town / District'] = df['City / Town / District'].apply(fill_missing)

df
```

```python
#replacing all NaN values in quesion with "Not Answered" as the person who filled the survey may have intentionally left that question so we cannot deduce what the answer could have been
# Example: Custom fill based on your knowledge
df.loc[df['Q4'].isnull(), 'Q4'] = 'Not Answered'
df.loc[df['Q5'].isnull(), 'Q5'] = 'Not Answered'
df.loc[df['Q6'].isnull(), 'Q6'] = 'Not Answered'
df.loc[df['Q7'].isnull(), 'Q7'] = 'Not Answered'
df.loc[df['Q8'].isnull(), 'Q8'] = 'Not Answered'
df.loc[df['Q9'].isnull(), 'Q9'] = 'Not Answered'
df.loc[df['Q10'].isnull(), 'Q10'] = 'Not Answered'
df.loc[df['Q11'].isnull(), 'Q11'] = 'Not Answered'
df.loc[df['Q12'].isnull(), 'Q12'] = 'Not Answered'
df.loc[df['Q15'].isnull(), 'Q15'] = 'Not Answered'
df
```

```python
import numpy as np
from datetime import datetime
for column in df.columns:
    # Strip leading and trailing whitespaces
    df[column] = df[column].astype(str).str.strip()
    
    # Replace multiple spaces with a single space
    df[column] = df[column].str.replace(r'\s+', ' ', regex=True)

# Changing '2 3' to 23 in Q3 column
df['Q3'] = df['Q3'].replace('2 3', 23)

# Getting the current year
current_year = datetime.now().year

# Convert '1993' to age assuming that the survey was filled out recently
def convert_year_to_age(value):
    if str(value).isdigit() and int(value) > 1900:
        return current_year - int(value)
    return value

df['Q3'] = df['Q3'].apply(convert_year_to_age)

# Convert Q3 to numeric after replacement
df['Q3'] = pd.to_numeric(df['Q3'], errors='coerce')

# Changing '6l' and '7O' to 61 and 70 in Q4 column
df['Q4'] = df['Q4'].replace('6l', 61)
df['Q4'] = df['Q4'].replace('7O', 70)

# Function to convert 5.2 to inches assuming it is 5 feet and 2 inches
def convert_specific_value(value):
    if value == '5.2':
        feet = 5
        inches = 2
        return feet * 12 + inches
    return value

# Apply the conversion function only to the specific value '5.2'
df['Q4'] = df['Q4'].apply(convert_specific_value)

# replacing particular values in the following columns to make data consistent
df['Q5'] = df['Q5'].replace('US', 'United States of Ameria')
df['Q5'] = df['Q5'].replace('china', 'China')
df['Q5'] = df['Q5'].replace('pakistan', 'Pakistan')
df['Q5'] = df['Q5'].replace('United States ; Pakistan', 'United States of America, Pakistan')
df['Q5'] = df['Q5'].replace('USA', 'United States of America')
df['Q5'] = df['Q5'].replace('United States', 'United States of America')
df['Q6'] = df['Q6'].replace(['economics', 'Economic', 'Economics'], 'Economics')
df['Q6'] = df['Q6'].replace(['Computer Science', 'Computer science'], 'Computer Science')
df['Q6'] = df['Q6'].replace(['Computer Science', 'Computer science'], 'Computer Science')


#removing duplicates if they are there
df = df.drop_duplicates()
df
```
